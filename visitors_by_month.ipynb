```python
import pandas as pd
from bokeh.plotting import figure, output_file, show, output_notebook
```

```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data/by_month.csv')
```

```python
# remove total column 
df.drop('Total', axis= 1, inplace=True)
```

```python
df
```

```python
months = list(df.columns)
#months
year = list(range(1992, 2014,1))
```

```python
df.info()
```

```python
df_new = pd.melt(df, id_vars = 'Year', var_name='Month', value_name = 'Visitors')
```

```python
df_new.sample(10)
```

```python
month_dict = {"Jan":1,"Feb":2,"Mar":3,"Apr":4, "May":5, "Jun":6,
     "July":7,"Aug":8,"Sep":9,"Oct":10,"Nov":11,"Dec":12}
```

```python
def change_month(x):
    return month_dict[x]
```

```python
df_new['Month'] = df_new['Month'].apply(change_month)
```

```python
df_new.sample(10)
```

```python
df_new['day'] = 1
```

```python
df_new['date'] = pd.to_datetime(df_new[['Year', 'Month', 'day']])
```

```python
data = df_new[['date', 'Visitors']]
```

```python
data.info()
```

```python
TOOLS = 'crosshair,save,pan,box_zoom,reset,wheel_zoom'
p = figure(title="No of visitors by Time", y_axis_type="linear",x_axis_type='datetime', tools = TOOLS)
```

```python
p.line(data['date'], data.Visitors, legend="visitors", line_color="purple", line_width = 3)
p.legend.location = "top_left"

p.xaxis.axis_label = 'Year'
p.yaxis.axis_label = 'number of visitors'

output_file("simple_plot.html", title="No of visitors in Nepal")

show(p)
```
